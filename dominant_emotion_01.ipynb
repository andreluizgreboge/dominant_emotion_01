```python
import cv2
import matplotlib.pyplot as plt
!pip install deepface
from deepface import DeepFace
```

```python
pics = ['a.png','b.jpg','c.jpg','d.jpg',
        'e.png','f.png','g.jpg','h.jpg',
        'i.jfif','j.jpg','k.jpg','l.jpeg']

for pic in pics:
  img1 = cv2.imread(pic)
  plt.imshow(img1[:,:,::-1])
  result = DeepFace.analyze(img1, actions=['emotion'], enforce_detection=False)
  emocao = result['dominant_emotion']
  plt.title('dominant emotion: '+emocao, fontsize=20)
  plt.axis('off')
  plt.show()
```
